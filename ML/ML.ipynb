### Primary feature engineering

- сгенерим дополнительные признаки из меты
- будем предсказывать главное настроение: EDA показал, что у 74% треков только 1 настроение

```python
def data_preprocessing(df_):

    df = df_.copy()

    df['release_date'] = pd.to_datetime(df['release_date'])
    df['release_year'] = df['release_date'].dt.year

    df['genre_cnt'] = df['genre'].apply(lambda x: len(x))
    df['genre'] = df['genre'].apply(lambda x: ', '.join(sorted(x)))

    df['instrument_cnt'] = df['instrument'].apply(lambda x: len(x))
    df['instrument'] = df['instrument'].apply(lambda x: ', '.join(sorted(x)))

    df.drop([
        'track_id',
        'artist_id',
        'album_id',
        'path',
        'mood/theme',
        'track_name',
        'release_date',
        'url',
        'mood_class',
        'situation_class',
        'for_summer',
        'for_children',
        'for_holiday',
        'for_sport',
        'for_christmas',
        'for_travel',
        'for_party',
        'is_sad_dark', 
        'is_energetic', 
        'is_positive', 
        'is_romantic', 
        'is_calm', 
        'is_deep_emotional'
    ], axis=1, inplace=True)

    return df
```

Выводы:
- Главная метрика — f1_macro, т.к. классы несбалансированы

- Ожидаемая ситуация: слабые границы между классами и сильная перекрываемость признаков -> получаем стабильно f1 в районе 0.3

    - уменьшить мультиколлинеарность: убрать дублирующие фичи (centroid/rolloff), агрегировать или применить PCA для признаков
    - отбирать признаки более продвинуто: RFE / Lasso / tree-based feature importance
    - перезвесить классы (class_weight / stratified sampling / SMOTE)
    - попробовать нелинейные модели: RandomForest / Catboost и тд
    - изучить важность признаков и их влияние на целевую переменную через SHAP/feature permutation

#### CatColumn Transformer

```python
cat_columns = [
    'genre', 'instrument', 'artist_name', 'album_name',
]
```

```python
class CatColumnsTransformer(BaseEstimator, TransformerMixin):

    """
    Класс для mean target encoding категориальных переменных в задаче мультикласса
    """
    
    def __init__(self, cat_columns):
        self.cat_columns = cat_columns
    
    def fit(self, X, y):

        df_with_y = pd.concat([X[self.cat_columns], y], axis=1)
        y_name = y.name
        
        df_with_y = pd.get_dummies(df_with_y, columns=[y_name], dtype=int)
        self.target_cols = [x for x in df_with_y.columns if 'major_mood_class_' in x]

        self.global_means_maps = df_with_y[self.target_cols].mean().to_dict()

        self.encode_maps = {}
        
        for col in self.cat_columns:
        
            dict_for_one_col = df_with_y.groupby(col)[self.target_cols].mean().to_dict(orient='index')
            self.encode_maps[col] = dict_for_one_col
        
        return self
    
    def transform(self, X):

        X_ = X.copy()
        res = []

        for col in self.cat_columns:

            m = X_[col].map(self.encode_maps[col])
            m = m.apply(
                lambda v: v if isinstance(v, dict) else self.global_means_maps
            )

            df_col = pd.DataFrame(m.tolist(), index=X_.index)
            df_col = df_col.fillna(self.global_means_maps)
            df_col.columns = [f"{col}_{c.split('_')[-1]}" for c in self.target_cols]
            
            res.append(df_col)

        X_ = X_.drop(columns=self.cat_columns)
        X_ = pd.concat([X_] + res, axis=1)
        
        return X_
```

```python
scaler = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), selector(dtype_include ='number')),                  
    ],
    remainder='passthrough',          
    verbose_feature_names_out=False 
).set_output(transform='pandas')
```

#### OneVsRestClassifier

```python
params_grid_lr = {
    'subset__columns': subsets_columns,
    'lr__estimator__penalty': ['l1', 'l2', 'elasticnet'],
    'lr__estimator__C': [0.001, 0.01, 0.5, 1, 10, 100],
    'lr__estimator__class_weight': ['balanced', None],
    'lr__estimator__solver': ['lbfgs', 'newton-cg'],
    #'lr__estimator__max_iter': [100, 500],
}, 
```

```python
pipe_ovr_lr = Pipeline([
    ('scaler', scaler), 
    ('mte', CatColumnsTransformer(cat_columns)), 
    ('subset', ColumnSelector()),             
    ('lr', OneVsRestClassifier(LogisticRegression(max_iter=1000)))
])
```
